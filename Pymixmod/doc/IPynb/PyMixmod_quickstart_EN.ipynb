# PyMixmod - quickstart

## Install PyMixmod

**PyMixmod** is available on **Linux - 64 bits** and **MacOSX** environments.

It is **Python 2.7.x** and **Python 3.6.x** compatible

### Installation on **Linux - 64 bits**

If **Anaconda** distribution is already installed on your workstation, you should skip to the 3rd step bellow, else:

1. Download the **Anaconda** installer from http://continuum.io/downloads site(you are free to choose between **Python 2.7** and  **Python 3.6** distribution)

2. Follow the guide on the distribution site for installing  **Anaconda**.

    > **Some notes** on *Anaconda installation (thereafter, <**CONDA_PATH**> is referencing the directory of this installation):
    >    * You are free to choose the <**CONDA_PATH**> value at the time of the Anaconda installation (provided you have enough privileges). The default value for <**CONDA_PATH**> (proposed by the installer) is :
    
    >        * **\$HOME/anaconda** if you choose **Python 2.7**
    >        * **\$HOME/anaconda3** if you choose  **Python 3.6**
    >    * The **Anaconda** installer puts in **<*CONDA_PATH*>/bin** many executables, including **conda** et **python**, which will be discussed later. 
    >    * The **Anaconda** installer asks, at the last installation step, if you want it to add **<*CONDA_PATH*>/bin** at your **PATH** environnement variable: 
    >> "Do you wish the installer to prepend the Anaconda install location to PATH in your /home/poli/.bashrc ? [yes|no]").

    > If you answered "no" to this question (and you could have good reasons to do it) you have to be sure that the executables used to perform the operations described bellow (**conda**, **python** etc.) are those provided in **<*CONDA_PATH*>/bin**

3. Download the **PyMixmod** package according to your **Anaconda** installation **Python** version:
    * *pymixmod-0.1-py27_0.tar.bz2* pour *Linux et Python 2.7*
    * *pymixmod-0.1-py36_0.tar.bz2* pour *Linux et Python 3.6* 

4. Install the previously downloaded package with the **conda** command provided in **<*CONDA_PATH*>/bin**:
    ```
    conda install pymixmod-0.1-pyXX_0.tar.bz2 # XX=27 or 36
    ```
    **NB:** If the path <*CONDA_PATH*>/bin is not available in your execution environment, then you have to execute, for instance:
    ```
    $HOME/anaconda/bin/conda install pymixmod-0.1-py27_0.tar.bz2
    ```


### Installation on **MacOSX**

Follow the same steps as for **Linux** installation by using at the **3rd** step the **MacOSX** packages:

* *OSX_pymixmod-0.1-py27_0.tar.bz2* pour *MacOSX et Python 2.7*
* *OSX_pymixmod-0.1-py36_0.tar.bz2* pour *MacOSX et Python 3.6*

## Using PyMixmod

### Preliminary remarks

After having successfully installed the package PyMixmod you should have access to:

* the API documentation at **<*CONDA_PATH*>/share/pymixmod/doc/html/index.html** (just open it with your browser)
* some examples in **<*CONDA_PATH*>/share/pymixmod/examples/** directory

##### Use the appropriate **python** binaries

In order to run the following samples, you have to use the binaries (**python**, **ipython**) which are part of  your **Anaconda** installation (and **NOT** those being part of your operating system distribution). For instance, if you have installed **Anaconda** in the location proposed by default, you should use:


* **$HOME/anaconda/bin/python** if you have chosen *Python 2.7*

* **$HOME/anaconda3/bin/python** if you have chosen *Python 3.4*

#### Portability *Python 2.7* - *Python 3.6*

The following examples work well in both **Python 2.7** and **Python 3.6** environnements.

In order to ensure the portability, the following imports at the beginning of each script or *Python* session are present in all given examples:

````python
from __future__ import absolute_import, division, print_function, unicode_literals
````

#### Using **Pandas**

Data manipulation in *Python* is greatly facilitated by the use of the *Pandas* library (see http://pandas.pydata.org/ for details).

*Pandas* allows, among others, the loading of **.csv** files as **DataFrame** objects, which are compatible with *PyMixmod* :

````python
import pandas as pd

# data = pd.read_csv("/path/to/datafile.csv")
````

#### Using *PyMixmod* package

Currently, the *PyMixmod* package includes an unique Python module called **mixmod** which must be imported before using it:

````python
import mixmod
````

In order to avoid the use in the scripts of long and complicated names for referencing frequently used constants, customarily one imports separately the **mixmod.gm** class containing enums, constants and various shortcuts:

````python
from mixmod import gm
````

For instance, in this way, one references **mixmod.gm.ModelName.Gaussian_p_Lk_Ck** as **gm.ModelName.Gaussian_p_Lk_Ck** or even (thanks to a shortcut) as **gm.Gaussian_p_Lk_Ck** which is even more concise.

### Example : Classification - quantitative data

#### Modules importation, data loading:

````python
from __future__ import absolute_import, division, print_function, unicode_literals
import pandas as pd
import mixmod
from mixmod import gm  # gm contains global constants (enum items etc.)

data = pd.read_csv("../../data/iris.train")
data = data.iloc[:, :4]  # in order to exclude the last column, containing unwanted labels
````

#### Classification computing, results displaying:

````python
cluster = mixmod.cluster(data, [2, 3, 4], gm.QUANTITATIVE)
print(cluster.summary())
````

#### Drawing the results with the *plot()* method

````python
# for details, see /doc/html/pymixmodapi.html#visualization-utilities
cluster.plot(y=["Sepal.Length", "Sepal.Width", "Petal.Length"])
````

#### Drawing histograms

````python
cluster.hist(variables=["Sepal.Length", "Sepal.Width"])
````

### Exemple : Classification - qualitative data

#### Modules importation, data loading:

````python
from __future__ import absolute_import, division, print_function, unicode_literals
import pandas as pd
import mixmod
from mixmod import gm  # gm contains global constants (enum items etc.)

data = pd.read_csv("../../data/car.csv", dtype="category")
````

#### Classification computing, results displaying:

````python
cluster = mixmod.cluster(data, 7, gm.QUALITATIVE)
print(cluster.summary())
````

#### Drawing the results with the *plot()* method

````python
cluster.plot()
````
