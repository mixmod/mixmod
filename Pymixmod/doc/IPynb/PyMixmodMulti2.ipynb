# Classification, données qualitatives (2)

Importations :

```python
import pandas as pd

import mixmod
from mixmod import gm
```

Chargement des données :

```python
data = pd.read_csv("../../data/car.csv", dtype="category")
```

```python
data
```

Classification :

```python
cluster = mixmod.cluster(data, 7, gm.QUALITATIVE)
```

L'objet résultat :

```python
cluster
```

Affichage Rmixmod-like :

```python
print(cluster)
```

La méthode **plot()** :

```python
fig = cluster.plot()
```
