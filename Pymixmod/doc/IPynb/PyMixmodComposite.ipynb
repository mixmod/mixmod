# Classification, données composites

Importations :

```python
import pandas as pd

import mixmod
from mixmod import gm
```

Chargement des données :

```python
data = pd.read_csv("../../data/heterodata.train")
```

Les variables qualitatives ont des valeurs entières :

```python
data["L"].dtype
```

On voudrait que les données qualitatives soient typées **category** (équivalent des **factors** dans **R**)
La fonction **set_categories()** permettra de faire cette conversion :

```python
def set_categories(df):
    for k in df.columns:
        if df[k].dtype.name == "int64":
            df[k] = df[k].astype("category")


# exécution:
set_categories(data)
data["L"].dtype
```

Classification :

```python
cluster = mixmod.cluster(data, 2, gm.HETEROGENEOUS)
```

Résultat :

```python
print(cluster)
```
